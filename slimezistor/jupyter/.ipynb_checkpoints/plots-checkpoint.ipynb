```python
# Plot cosine at audio frequency

import matplotlib.pyplot as plt
import numpy as np

f_audio = 150 # Hz
sample_rate = 100*f_audio #samples per second
sample_time = 0.1 # seconds

# time values

t_values = np.linspace(0, sample_time,int(sample_rate*sample_time))
v_values = np.cos(2*np.pi*f_audio*t_values)
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(6)

#v_values = t_values + t_values
                       
plt.plot(t_values, v_values,"+")
    
plt.xlabel('t [s]')
plt.ylabel('$\cos{(2\pi f_{audio} t)}$')
plt.title('Plot of $\cos{(2\pi f_{audio} t)}$ vs t')
#plt.ylim(0, 1.5)
plt.show()
```

```python
# Plot cosine at audio frequency

import matplotlib.pyplot as plt
import numpy as np

f_audio = 150 # Hz
sample_rate = 100*f_audio #samples per second
sample_time = 0.1 # seconds

# time values

t_values = np.linspace(0, sample_time,int(sample_rate*sample_time))
v_values = np.sign(np.cos(2*np.pi*f_audio*t_values))
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(6)

#v_values = t_values + t_values
                       
plt.plot(t_values, v_values,"+")
    
plt.xlabel('t [s]')
plt.ylabel('$\cos{(2\pi f_{audio} t)}$')
plt.title('Plot of $\cos{(2\pi f_{audio} t)}$ vs t')
#plt.ylim(0, 1.5)
plt.show()
```

```python
# Plot ramp 

import matplotlib.pyplot as plt
import numpy as np

f_audio = 440 # Hz



sample_rate = 100*f_audio #samples per second
sample_time = 0.15 # seconds

C = .47e-6 # farads
R = 5000 # ohms
tau = R*C

t_sweep = 0.01 # seconds
n_sweep = int(t_sweep*sample_rate) # number of points in a sweep

t_values = np.linspace(0, sample_time,int(sample_rate*sample_time))

v_values = (t_values%n_sweep)
v_values = ((t_values*sample_rate)%n_sweep)/n_sweep


boolean_rise = 0.5*(np.sign(np.sin(2*np.pi*(0.5/t_sweep)*t_values)) + 1)
boolean_fall = -0.5*(np.sign(np.sin(2*np.pi*(0.5/t_sweep)*t_values)) - 1)
                       
t_values_rise = boolean_rise*v_values
t_values_fall = boolean_fall*v_values

v_values_rise = boolean_rise*(1 + np.exp(-t_sweep/tau) - np.exp(-t_sweep*t_values_rise/tau))
v_values_fall = boolean_fall*(np.exp(-t_sweep*t_values_fall/tau))

v_values_total = v_values_rise + v_values_fall

f = plt.figure()
f.set_figwidth(20)
f.set_figheight(6)

#v_values = t_values + t_values
                       
plt.plot(t_values,v_values_total,"+")
    
plt.xlabel('t [s]')
plt.ylabel('capacitor voltage[normalized units]')
plt.title('Plot of voltage on capacitor vs t')
#plt.ylim(0, 1.5)
plt.show()
```
